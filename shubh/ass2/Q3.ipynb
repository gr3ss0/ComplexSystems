```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def entropy_function(p):
    if (p*(1-p) != 0): 
        
        value = -(p*math.log(p, 2) + (1-p)* math.log(1-p, 2))
    else:
        value=0 #entropy should be 0 if p is a root of the polynomial x-x^2
    return value
    
```

```python
p_values=np.linspace(0, 1, 500)
```

```python
entropy_values=np.array([entropy_function(p) for p in p_values])
```

```python
plt.plot(p_values, entropy_values)
plt.axvline(x=0.5, linestyle='--', color='red', label='0.5 is the maximizer')
plt.title('Entropy change with different values of p')
plt.xlabel('Probability of heads(p)')
plt.ylabel('Entropy in bits')
plt.legend(loc='center left')
plt.savefig('entropy_plot.png')
plt.show()
```
