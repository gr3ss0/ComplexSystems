```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import math
```

```python
np.random.seed(42)
```

```python
G_er=nx.Graph()
```

```python
nodes=np.array(range(150))
nodes_str=[str(x) for x in nodes]
G_er.add_nodes_from(nodes_str)
print(G_er)
```

```python
from itertools import combinations
```

```python
for (u, v) in list(combinations(nodes_str, 2)):
    r=np.random.uniform(0,1)
    if r < 0.08:
        G_er.add_edge(u, v)
#now we have obtained the Erdos Renyi Graph
```

```python

positions = nx.spring_layout(G_er)#this is saying where the points are in 2d plane, should use circular layout and spring layout and check which works better



#nx.draw_networkx_nodes(G_er, positions, node_color='lightblue') #this draws the vertices

nx.draw_networkx_edges(G_er, positions, width=2) #this draws edges

#nx.draw_networkx_labels(G_er, positions, font_color='black') #this is the node labels, no need to specify manually


plt.axis('off')

plt.show()
```

```python
#number and size of connected components

num_connected_comp=nx.number_connected_components(G_er)
print(num_connected_comp) #connected graph
connected_comp=nx.connected_components(G_er)
list_connected_comp=list(connected_comp) #this will be a list of sets, each set is a connected component

size_list=[len(c) for c in list_connected_comp] #capture the length of each set in a list
print(size_list)
```

```python
deg_dict=dict(G_er.degree())
print(deg_dict)
plt.hist(deg_dict.values())
plt.title('Degree Histogram')
plt.xlabel('Degree')
plt.ylabel('Counts')
plt.savefig('hist_deg.png')
plt.show()
```

```python


closeness = dict(nx.closeness_centrality(G_er))
print(closeness.values())#getting all the values
plt.hist(closeness.values())
plt.title('Closeness Centrality')
plt.xlabel('Closeness Centrality')
plt.ylabel('Counts')
plt.savefig('closeness.png')
plt.show()
```

```python
betweenness=dict(nx.betweenness_centrality(G_er))
print(betweenness)

plt.hist(betweenness.values())
plt.title('Betweenness Centrality')
plt.xlabel('Betweenness Centrality')
plt.ylabel('Counts')
plt.savefig('betweenness.png')
plt.show()
```

```python
#normalized degree histogram with overlay of poisson
from scipy.stats import poisson
n=150
p=0.08
lam=n*p
deg_dict=dict(G_er.degree())
degree_values=list(deg_dict.values())
degree_range=np.arange(min(degree_values)-1, max(degree_values)+1)
poisson_list=np.array([poisson.pmf(x, lam) for x in degree_range])
plt.plot(degree_range, poisson_list, 'ro-')
plt.hist(deg_dict.values(), density=True, bins=20)
plt.title('Degree Histogram-Normalized: 20 bins')
plt.xlabel('Degree')
plt.ylabel('Normalized Counts')
plt.savefig('hist_deg_normalized_bins20.png')
plt.show()
```

```python
plt.plot(degree_range, poisson_list, 'ro-')
plt.hist(deg_dict.values(), density=True, bins=10)
plt.title('Degree Histogram-Normalized:10 bins')
plt.xlabel('Degree')
plt.ylabel('Normalized Counts')
plt.savefig('hist_deg_normalized_bins10.png')
plt.show()
```
